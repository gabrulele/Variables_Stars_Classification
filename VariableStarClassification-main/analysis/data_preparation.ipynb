## Data analysis

```python
import pandas as pd
import os

# Ottieni la directory padre della directory corrente
parent_dir = os.path.dirname(os.getcwd())

local_file_path  = "\data\PLV_LINEAR.csv"

fullDataSetPath = parent_dir + local_file_path



data = pd.read_csv(fullDataSetPath, sep=";")

# Carica solo le prime 5 righe per diagnosticare la struttura
data.head()
```

## Data cleaning

```python
data.isnull().sum() # il dataset non presenta valori mancanti
```

```python
data = data.drop_duplicates()  # Rimuove righe duplicate
data = data.drop("#", axis=1)  # Rimuove la colonna di valri NaN
data.head()
```

```python
# Conta le occorrenze per ciascun valore nella colonna 'LCtype'
conteggi = data['LCtype'].value_counts()
print(conteggi)
```

```python
'''
Osserviamo che le classi 3, 9, 11, 8 sono scarsamente popolate, possono essere rimosse

'''

# Lista dei valori da rimuovere come numeri interi
valori_da_rimuovere = [3, 9, 11, 8]  # Rimuovi le virgolette per avere numeri interi

# Filtra il dataframe escludendo questi valori
data = data[~data['LCtype'].isin(valori_da_rimuovere)]

# Conta le occorrenze per ciascun valore nella colonna 'LCtype'
conteggi = data['LCtype'].value_counts()
print(conteggi)
```

## Salvataggio file csv
Dopo la fase di data_cleaning il file pu√≤ essere scaricato e utilizzato per il training del modello

```python
import os

# Ottieni la directory corrente
currDir = os.getcwd()

# Ottieni la directory padre
parent_dir = os.path.dirname(currDir)

# Costruisci il percorso completo utilizzando os.path.join
fullOutputPath = os.path.join(parent_dir, "data", "output", "output.csv")

# Stampa il percorso per verificare che sia corretto
print(fullOutputPath)

data.to_csv(fullOutputPath , index=False)

```
